## Quick math note: Random walk
* Key time series generator
* Important for asset prices
$$y_t = y_{t-1} + e_t$$

## Plotting a random walk

```python
y = np.random.standard_normal(100).cumsum()
# cumsum():
#   1st time: e(0)
#   2nd time: e(0)+e(1)
#   3rd time: e(0)+e(1)+e(2)
plt.plot(y)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Random walk level')
plt.title('A Random Walk')
```

## Two random walks

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0) 
# generate a variable ()20 rows 2 columns
# axis=0: means that I will use column to calculate the "cumsum"
# 20 rows 2 columnsso there are 2 random walks
plt.plot(y)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Random walk level')
plt.title('Two Random Walks')
plt.legend(('A','B'))
#plt.legend()
```

## Two random walks again

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
plt.plot(y[:,0],'b-', lw=1.5,label='Winners')  # blue, solid
plt.plot(y[:,1],'r--',lw=1.5,label='Losers')  # red, dash
plt.grid()
plt.xlabel('Time')
plt.ylabel('Random walk level')
plt.title('Two Random Walks')
plt.legend()
```

## Plotting with two axes
* Sometimes your data doesn't have the same units, but you want to compare it
* You can have two y-axes for this

```python
y = np.random.standard_normal(50).cumsum(axis=0)
z = 100.*np.random.standard_normal(50).cumsum(axis=0)
# Try to delete "axis=0", it turned out to be the same
y = np.random.standard_normal(50).cumsum()
z = 100.*np.random.standard_normal(50).cumsum()
fig, ax1 = plt.subplots()  # how to understand??
plt.plot(y,'b-', lw=1.5,label='Winners')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Winner level')
plt.title('Two Random Walks')
ax2 = ax1.twinx() # how to understand the x here?
plt.plot(z,'r--',lw=1.5,label='Losers')
plt.ylabel('Random walk level')
```

## Setting a legend is a little tricky

```python
y = np.random.standard_normal(50).cumsum(axis=0)
z = 100.*np.random.standard_normal(50).cumsum(axis=0)
fig, ax1 = plt.subplots()
plt.plot(y,'b-', lw=1.5,label='Winners')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Winner level')
plt.title('Two Random Walks')
ax2 = ax1.twinx()
plt.plot(z,'r--',lw=1.5,label='Losers')
plt.ylabel('Random walk level')
plt.legend()
```

```python
y = np.random.standard_normal(50).cumsum(axis=0)
z = 100.*np.random.standard_normal(50).cumsum(axis=0)
fig, ax1 = plt.subplots()
ln1 = ax1.plot(y,'b-', lw=1.5,label='Winners')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Winner level')
plt.title('Two Random Walks')
ax2 = ax1.twinx()
ln2 = ax2.plot(z,'r--',lw=1.5,label='Losers')
plt.ylabel('Random walk level')
lns = ln1 + ln2 # this is a list of lines
plt.legend(lns,['winners','losers'])
```

## Multiple plot panels

```python
y = np.random.standard_normal(50).cumsum(axis=0)
z = 100.*np.random.standard_normal(50).cumsum(axis=0)
# top panel )(2x1 grid)
plt.subplot(211) # 2: 2 rows, 1: 1 column, 1: the top one #What if the number > 10?
plt.plot(y,'b-', lw=1.5,label='Winners')
plt.grid()
plt.ylabel('Winners')
plt.title('Random Walk Levels')
# bottom panel
plt.subplot(212)
plt.plot(z,'r--',lw=1.5,label='Losers')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Losers')
```

## Histogram

```python
# basic histogram
y = np.random.standard_normal(250)
plt.hist(y)
plt.grid()
```

## Multiple series histogram
* Note:  Try this with multiple panels too

```python
# Two series
y = np.random.standard_normal(5000)
z = np.random.uniform(low=-3.,high=3.,size=5000)
plt.hist((y,z),color=['blue','red'],label=['Winner','Loser'],bins=25)
plt.grid()
plt.legend()
```
